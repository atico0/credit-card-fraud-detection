```python
df.drop('Time', inplace=True, axis=1)
df.head()
```

##Padronizando os dados

```python
from sklearn.preprocessing  import StandardScaler
```

```python
x_valores = df.iloc[:,:-1].values
y_valores = df.iloc[:,-1].values
x_valores = StandardScaler().fit_transform(x_valores)
x_valores
```

###Fazendo a sobreamostragem

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy='minority')
x_over, y_over = smote.fit_resample(x_valores, y_valores)
```

```python
np.unique(y_over, return_counts=True)
```

```python
plt.hist(y_over)
```

###separando os dados entre treino e teste

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_over, y_over, test_size=0.25)
```

##Treinando modelos

###Naive bayes

```python
from sklearn.naive_bayes import GaussianNB
from yellowbrick.classifier import ConfusionMatrix
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score ,ConfusionMatrixDisplay
```

```python
naive = GaussianNB() #criando e treinando algoritmo de naive bayes
naive.fit(x_train ,y_train)
```

```python
pred = naive.predict(x_test)
```

```python
print(classification_report(pred,y_test))
```

```python
accuracy_score(pred,y_test)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test,pred)).plot(colorbar=True)
```

##Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(criterion='entropy')
tree.fit(x_train, y_train)
```

```python
pred = tree.predict(x_test)
```

```python
print(classification_report(pred,y_test))
```

```python
accuracy_score(pred,y_test)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test,pred)).plot(cmap='crest')
```

###redes neurais

```python
from sklearn.neural_network import MLPClassifier
```

```python
rede_neural = MLPClassifier(max_iter=1000, verbose=True, solver='adam', tol=0.0000100)
rede_neural.fit(x_train, y_train)
```

```python
pred = rede_neural.predict(x_test)
```

```python
print(classification_report(pred,y_test))
```

```python
accuracy_score(pred,y_test)
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test,pred)).plot(cmap=sns.cubehelix_palette(as_cmap=True))
```
